```python
import pandas as pd
```

```python
outdf = pd.read_csv('angle_stats_datasource.csv')
outdf
```

```python
ligtypes = outdf.geotype_label.value_counts().index.values # Pull out all labelled ligtypes
```

```python
ligtypes
```

```python
stat_columns = [x for x in outdf.columns.values if 'angle' in x] # Pull out angle features
```

```python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95): # Calculate mean confidence intervals
    n = len(data)
    m, se = np.mean(data), np.std(data)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, h
```

```python
newdflist = []
for t in ligtypes:
    tdf = outdf[outdf.geotype_label == t]
    outdict = {'ligtype':t,'denticity':int(np.mean(tdf.cn.values))}
    for angle in stat_columns:
        vals = tdf.loc[:,angle].values
        mean, conf = mean_confidence_interval(vals) # Calculate 95% confidence intervals
        outdict.update({angle+'_mean':mean,angle+'_95conf':conf})
    newdflist.append(outdict)
```

```python
statsdf = pd.DataFrame(newdflist)
```

```python
statsdf.to_csv('ligtype_angle_reference.csv',index=False) # Use this dataframe as reference for ligand geometry types
```
